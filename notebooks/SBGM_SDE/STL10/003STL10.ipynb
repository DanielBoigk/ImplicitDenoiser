```julia
using DataDeps
using Downloads

# 1. Register the STL-10 Data Dependency
register(DataDep(
    "STL-10",
    """
    Dataset: The STL-10 dataset for unsupervised feature learning.
    Website: https://cs.stanford.edu/~acoates/stl10/
    """,
    # We fetch the official binary version
    "https://cs.stanford.edu/~acoates/stl10/stl10_binary.tar.gz",
    # SHA256 checksum ensures the download is safe and unmodified
    "4d44bca6f28da3da4387cc3993f40f0c05943fcfd6bfa3f34563ff44f24300e8",
    post_fetch_method = unpack # Automatically extracts the .tar.gz
))
```

```julia
function load_stl_bin(filepath, num_images)
    open(filepath, "r") do io
        # 96 width * 96 height * 3 channels = 27648 bytes per image
        raw_data = read(io, 96 * 96 * 3 * num_images)
        # Reshape to (Width, Height, Channels, NumImages)
        return reshape(raw_data, (96, 96, 3, num_images))
    end
end

function load_stl_labels(filepath, num_images)
    open(filepath, "r") do io
        return Int.(read(io, num_images)) # Labels are 1-10
    end
end

# 2. Trigger the automated download/path resolution
# This returns the local folder path where DataDeps extracted the files
@info "Resolving STL-10 dataset dependency..."
stl10_path = datadep"STL-10" 
inner_dir = joinpath(stl10_path, "stl10_binary")
```
